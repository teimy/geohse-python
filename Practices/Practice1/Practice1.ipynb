### Переменные

- Нужны для хранения информации
- Бывают разные типов:
    - `int` или `integer`: для хранения целых чисел
    - `bool`: для хранения `True` или `False`
    - `float`: для хранения нецелых чисел
    - `string`: для хранения строк (символов или слов)
    - Других типов
- Хранятся в памяти `python` если инициализированы (выполнен код, где определяется переменная)

````python
a = 3
b = True
c = 3.14
d = 'Доктор это вам'
````

### Операции над переменными

- Основные операторы:
    - `+`
    - `-`
    - `/`
    - `*`
- Операторы сравнения (как результат возвращают тип `bool`):
    -  `>` — больше, `>=` — больше или равно
    -  `<` — меньше, `<=` — меньше или равно
    -  `==` — проверка равенства (возрващает два если значение одинаково)
    -  `!=` — не равно
- Логические операторы:
    - `and`
    - `or`
    - `not`
- **Дополнительные операторы**:
    - `x**n` — `x` в степени `n` 
    - `//` — деление нацело
    - `%` — остаток от деления

````python
print('10 + 2 =', 10 + 2)
print('10 - 2 =', 10 - 2)
print('10/2 =', 10/2) # обратите внимание что результат уже не integer
print('10*2 =', 10*2)
print('10 > 2 =', 10 > 2)
print('10 < 2 =', 10 < 2)
print('10 == 2 =', 10 == 2)
print('10 != 2 =', 10 != 2)
print('True and False:', True and False)
print('True or False:', True or False)
print('not True:', not True)
print('2**3 =', 2**3)
print('9 // 4 =', 9 // 4)
print('9 % 4 =', 9 % 4)
````

### Условные операторы:

- оператор `if` позволяет выполнять блок кода при выполнении условия
- следующая конструкция:
    - `if` + условие + :
    - (отступ) Блок кода
    - Блок кода вне отступа выполняется без условий (при отсутствии блока `else`)
- `else`: блок кода, который будет выполняться если условие не выполнилось

````python
number = 2 # можно поменять число

if number >= 5:
    print('Число больше или равно пяти')
else:
    print('Число меньше пяти')
````

- `elif`: промежуточный блок, который позволяет определить дополнительное условие

````python
number = 3 # измените число так, чтобы выполнилась ветка `elif`

if number >= 5:
    print('Число больше или равно пяти')
elif number < 0:
    print('Число меньше нуля')
else:
    print('Число меньше пяти, но неотрицательное')
````
